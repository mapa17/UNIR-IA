```python
# Load image, convert it to grayscale, normalize it
image = skimage.color.rgb2gray(data.astronaut())
image = image/image.max()
fig = plt.figure(figsize= fig_size) #create an empty figure to plot into with 20x20 size
io.imshow(image)
plt.show()
```

# Ruido
 Crear una funcin que modele la adicin, de forma aleatoria, de artefactos impulsivos. El resultado tendr que apreciarse en la visualizacin de la imagen, que deber estar afectada por ruido de tipo sal y pimienta. Como se ha visto en la asignatura, estos artefactos toman valores de intensidad mximos o mnimos y afectan, aleatoriamente, a los pxeles de la imagen.

* La funcin a implementar debe aceptar la imagen original y devolver la imagen afectada por el ruido. Adems, sera deseable que aceptara un argumento adicional para indicar el porcentaje de pxeles que se vern afectados por estos artefactos.


```python
def sp_noise(image, p, seed=42):
    np.random.seed(seed)
    N = int(image.size * max(min(p, 1.0), 0.0) * 0.5)
    
    noisy_image = np.copy(image)
    # Add salt and pepper
    noisy_image.flat[np.random.randint(noisy_image.size, size=N)] = 1.0
    noisy_image.flat[np.random.randint(noisy_image.size, size=N)] = 0.0
    
    return noisy_image
```

```python
# Add 1% of salt and pepper noise
sp_image = sp_noise(image, 0.01)
fig = plt.figure(figsize=fig_size)
io.imshow(sp_image)
plt.show()
```

# Filtrar ruido
 A partir de la imagen ruidosa, buscaremos eliminar los artefactos impulsivos para el posterior tratamiento de la imagen. Para ello, se deber aplicar un filtro adecuado para este tipo de ruido.


```python
# Apply a median filter to get rid of s&p
median_image = skimage.filters.median(sp_image)
```

```python
fig = plt.figure(figsize=fig_size)
io.imshow(median_image)
plt.show()
```

# Caculate outlines
Por ltimo, a partir de la imagen obtenida en la etapa anterior, se busca identificar las siluetas de las estructuras en ella presentes. A este fin, se debe identificar y razonar qu tipo de operador corresponde aplicar.

```python
# Apply a sobel filter to detect edge intensity
sobel_image = skimage.filters.sobel(median_image)
fig = plt.figure(figsize=fig_size)
io.imshow(sobel_image)
plt.show()
```
