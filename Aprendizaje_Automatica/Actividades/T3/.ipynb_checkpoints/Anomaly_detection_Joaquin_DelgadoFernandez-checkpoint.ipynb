```python
IsolationInstances = [
    ("IsolationForest, n_estimators='100'",
        IsolationForest(n_estimators=100,
                               n_jobs = -1 , behaviour="new"
                               )),
    
    ("IsolationForest, n_estimators='200'",
          IsolationForest(n_estimators=200,
                               n_jobs = -1 , behaviour="new"
                              )),


    ("IsolationForest, n_estimators=300",
         IsolationForest(n_estimators=300,
                               n_jobs = -1 , behaviour="new"
                               ))
]
```

```python
scores = {}
```

```python
for label, clf in IsolationInstances:
    print(label)
    for i in [0.1,0.2,0.5]:
        clf.set_params(contamination =i)
        clf.fit(bcw)
        y_pred_train = clf.predict(bcw)
        
        mae = r2_score(test_data.diagnosis,y_pred_train)
        print('contamination '+str(i) + " -> " +str(mae))
        scores[label  + '  contamination '+str(i) + " -> "] = mae
```
